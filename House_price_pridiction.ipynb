```python
df3 = df2
```

```python
data = df3.drop(columns = ['Parking','Transaction','Type','Sqft_per_bed','Per_Sqft_Price'])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
# col_trans = make_column_transformer((OneHotEncoder(sparse_output = False),['Locality','Furnishing']),remainder='passthrough')

col_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['Locality', 'Furnishing']),
    remainder='passthrough',
    force_int_remainder_cols=False
)
```

```python
lnr = LinearRegression()
scal = StandardScaler()
model = make_pipeline(col_trans,scal,lnr)
```

```python
Input_data = data.drop(columns = ['Price'])
Output_data = data['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(Input_data, Output_data,test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
input =pd.DataFrame([[1050.0,3,2.0,'Semi-Furnished','Chittaranjan Park']],columns =['Area','BHK','Bathroom','Furnishing','Locality'])
```

```python
model.predict(input)
```
